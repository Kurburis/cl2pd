# Step 1: first download of data
---

At this stage, you will download a first amount of data, let's say 10 minutes, for the variables of your choice. In our case, we need the bunch-by-bunch intensity and luminosity from ATLAS and CMS

```python
# Defintion of the variables

FBCT_VAR = list(variablesDF.LHC[(variablesDF.LHC['Description'].str.contains('Main FBCT')) & (variablesDF.LHC['Tag']=='Intensity')]['Variable'])
LUMI_VAR = list(variablesDF.LHC[variablesDF.LHC['Tag']=='Luminosity']['Variable'])
CURRENT_VAR = list(variablesDF.LHC[variablesDF.LHC['Description']=='Wire current']['Variable'])
VAR = FBCT_VAR+LUMI_VAR+CURRENT_VAR
```

```python
# Time stamps

TZ_GVA = 'CET'
#T_NOW = pd.Timestamp.now(tz = TZ_GVA) #to work online
T_NOW = pd.Timestamp('2017-11-29 22:00',tz = TZ_GVA) #to work in MD4 2017
DELTA_T = 10. # number of initial minutes
T_INIT = T_NOW - pd.Timedelta(str(DELTA_T)+'m')

RAW_DATA = importData.cals2pd(VAR,T_INIT,T_NOW)
```

# Step 2: Organize and post process the data
---

One wants to organise the data in a dotdict with the following structure: myData.BEAM.BUNCH. The leave of this tree is a pandas DF containing resampled data for intensity and the luminosity, but also our final observable here: the effective cross-section. 

```python
# Filling scheme definition
# Please note that this tool is not meant to be used for operational filling schemes

FILL_PATTERN = getPresentFillingPattern(RAW_DATA['LHC.BCTFR.A6R4.B1:BUNCH_INTENSITY'],RAW_DATA['LHC.BCTFR.A6R4.B2:BUNCH_INTENSITY'])

print('==== FILLING PATTERN B1 ====')
print(FILL_PATTERN.atB1)
print('==== FILLING PATTERN B2 ====')
print(FILL_PATTERN.atB2)
```

```python
# Extract bbb data

# For fbct
FBCT_B1 = MDanalysis.raw2bbb(RAW_DATA['LHC.BCTFR.A6R4.B1:BUNCH_INTENSITY'],fill=FILL_PATTERN.atB1)
FBCT_B2 = MDanalysis.raw2bbb(RAW_DATA['LHC.BCTFR.A6R4.B2:BUNCH_INTENSITY'],fill=FILL_PATTERN.atB2)

#For lumi
LUMI_CMS = MDanalysis.raw2bbb(RAW_DATA['CMS:BUNCH_LUMI_INST'],fill=np.concatenate([FILL_PATTERN.atB1,FILL_PATTERN.atB2]))*1e30     #Problem of unit to be adressed
LUMI_ATLAS = MDanalysis.raw2bbb(RAW_DATA['ATLAS:BUNCH_LUMI_INST'],fill=np.concatenate([FILL_PATTERN.atB1,FILL_PATTERN.atB2]))*1e27 #Problem of unit to be adressed

# Wires Current

CURRENT_DF = RAW_DATA[CURRENT_VAR].interpolate(method='linear')
```

```python
# Finally our structure

sample_min = 60

myB1 = dotdict({})
myB2 = dotdict({})

for i in FILL_PATTERN.atB1:
    myDF = pd.DataFrame()
    myDF['Diff_Intensity'] = FBCT_B1[i].resample(str(sample_min)+'s').mean().diff()/float(sample_min)
    myDF['Lumi_ATLAS'] = LUMI_ATLAS[i].resample(str(sample_min)+'s').mean()
    myDF['Lumi_CMS'] = LUMI_CMS[i].resample(str(sample_min)+'s').mean()
    myDF = myDF.interpolate(method='linear')
    totLumi_mb = (myDF['Lumi_ATLAS'].values+myDF['Lumi_CMS'].values)*1e-27
    dI_dt = myDF['Diff_Intensity'].values
    myDF['Eff_section_mb'] = -dI_dt/totLumi_mb
    myB1['at_b'+str(i)]=myDF
    
for i in FILL_PATTERN.atB2:
    myDF = pd.DataFrame()
    myDF['Diff_Intensity'] = FBCT_B2[i].resample(str(sample_min)+'s').mean().diff()/float(sample_min)
    myDF['Lumi_ATLAS'] = LUMI_ATLAS[i].resample(str(sample_min)+'s').mean()
    myDF['Lumi_CMS'] = LUMI_CMS[i].resample(str(sample_min)+'s').mean()
    myDF = myDF.interpolate(method='linear')
    totLumi_mb = (myDF['Lumi_ATLAS'].values+myDF['Lumi_CMS'].values)*1e-27
    dI_dt = myDF['Diff_Intensity'].values
    myDF['Eff_section_mb'] = -dI_dt/totLumi_mb
    myB2['at_b'+str(i)]=myDF
    
myData = dotdict({'atB1':myB1,'atB2':myB2})

# Note that those lines are wrapped in a function for the next step, this cell has to be run only once.
```

# Step 3: re-download, re-post process
---

Being an online monitoring, one has now to download live data, to add them to the raw_data DF, to post process it, and to add them to the tree. 

```python
# Time definition

T1 = RAW_DATA.index[-1] # we start where we stopped
T2 = pd.Timestamp.now(tz=TZ_GVA) # up to now! 
if 1: # Or if u prefer to use a given time interval, just replace the 0 by 1 :) 
    DELTA_T = 10.
    T2 = T1 + pd.Timedelta(str(DELTA_T)+'m')

# Dowload the new DF

NEW_DATA = importData.cals2pd(VAR,T1,T2)

# Add it to the main DF

RAW_DATA=pd.concat([RAW_DATA,NEW_DATA])

# Post process/concat

myData = refreshData(FILL_PATTERN, NEW_DATA, myData, sample_min)
CURRENT_DF = pd.concat([CURRENT_DF,NEW_DATA[CURRENT_VAR].interpolate(method='linear')])
```

# Step 4: LET'S PLOT :) 

```python
# We want to plot B2 

fig = plt.figure(figsize=(20,10))
ax1 = fig.add_subplot(211)
for i in FILL_PATTERN.atB2:
    x=myData.atB2['at_b'+str(i)].index.tz_convert('CET')
    y=myData.atB2['at_b'+str(i)]['Eff_section_mb'].values
    plt.plot(x,y,'o-',label='b_'+str(i))
ax1.set_ylim(0,400)
plt.grid(ls=':')

plt.title('Machine Development 3263 - 15th June 2018')
ax1.set_ylabel('Eff. Cross-Section [mb]')
plt.plot(plt.xlim(),[80,80],'--k',label='BO limit')
plt.legend(loc='best',frameon=True)
plotFunctions.setSourcePlot(plt.gca(),mySource,color='k')

ax2 = fig.add_subplot(212, sharex=ax1)
x = CURRENT_DF.index.tz_convert('CET')
for i in CURRENT_DF.columns:
    y = CURRENT_DF[i].values
    ax2.plot(x,y,label=i)
ax2.set_xlabel('Wire Current [A]')
ax2.set_ylim([0,360])
plotFunctions.setSourcePlot(plt.gca(),mySource,color='k')
plt.legend(loc='best',frameon=True)
plt.grid(ls=':')
plt.xlim(pd.Timestamp(RAW_DATA.index[-1] - pd.Timedelta('1H'),tz=TZ_GVA).to_pydatetime(),pd.Timestamp(RAW_DATA.index[-1],tz=TZ_GVA).to_pydatetime())
plt.xlabel('Local Time [hh:mm]');
```

# Functions section
---

Here are all the function that, after discussion, will be pushed in a package, or will live inside this notebook. 

```python
def getPresentFillingPattern(FBCT_B1,FBCT_B2):
    '''
    From pandas series FBCT_BX, return a dotdict containing the present filling scheme in the machine. 
    '''
    fillB1 = np.where(FBCT_B1.dropna().iloc[0])[0]
    fillB2 = np.where(FBCT_B2.dropna().iloc[0])[0]
    return dotdict({'atB1':fillB1,'atB2':fillB2})
```

```python
def refreshData(FILL_PATTERN, NEW_DATA, myData, sample_min):
    
    # For fbct
    FBCT_B1 = MDanalysis.raw2bbb(NEW_DATA['LHC.BCTFR.A6R4.B1:BUNCH_INTENSITY'],fill=FILL_PATTERN.atB1)
    FBCT_B2 = MDanalysis.raw2bbb(NEW_DATA['LHC.BCTFR.A6R4.B2:BUNCH_INTENSITY'],fill=FILL_PATTERN.atB2)

    #For lumi
    LUMI_CMS = MDanalysis.raw2bbb(NEW_DATA['CMS:BUNCH_LUMI_INST'],fill=np.concatenate([FILL_PATTERN.atB1,FILL_PATTERN.atB2]))*1e30     #Problem of unit to be adressed
    LUMI_ATLAS = MDanalysis.raw2bbb(NEW_DATA['ATLAS:BUNCH_LUMI_INST'],fill=np.concatenate([FILL_PATTERN.atB1,FILL_PATTERN.atB2]))*1e27 #Problem of unit to be adressed
    
    for i in FILL_PATTERN.atB1:
        myDF = pd.DataFrame()
        myDF['Diff_Intensity'] = FBCT_B1[i].resample(str(sample_min)+'s').mean().diff()/float(sample_min)
        myDF['Lumi_ATLAS'] = LUMI_ATLAS[i].resample(str(sample_min)+'s').mean()
        myDF['Lumi_CMS'] = LUMI_CMS[i].resample(str(sample_min)+'s').mean()
        myDF = myDF.interpolate(method='linear')
        totLumi_mb = (myDF['Lumi_ATLAS'].values+myDF['Lumi_CMS'].values)*1e-27
        dI_dt = myDF['Diff_Intensity'].values
        myDF['Eff_section_mb'] = -dI_dt/totLumi_mb
        myData.atB1['at_b'+str(i)] = pd.concat([myData.atB1['at_b'+str(i)],myDF]).interpolate(method='linear')
    
    for i in FILL_PATTERN.atB2:
        myDF = pd.DataFrame()
        myDF['Diff_Intensity'] = FBCT_B2[i].resample(str(sample_min)+'s').mean().diff()/float(sample_min)
        myDF['Lumi_ATLAS'] = LUMI_ATLAS[i].resample(str(sample_min)+'s').mean()
        myDF['Lumi_CMS'] = LUMI_CMS[i].resample(str(sample_min)+'s').mean()
        myDF = myDF.interpolate(method='linear')
        totLumi_mb = (myDF['Lumi_ATLAS'].values+myDF['Lumi_CMS'].values)*1e-27
        dI_dt = myDF['Diff_Intensity'].values
        myDF['Eff_section_mb'] = -dI_dt/totLumi_mb
        myData.atB2['at_b'+str(i)] = pd.concat([myData.atB2['at_b'+str(i)],myDF]).interpolate(method='linear')

    return myData
    
```
